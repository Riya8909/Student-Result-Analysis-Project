```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head(5))
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop unnamed column

```python


df = df.drop("Unnamed: 0", axis = 1)
print(df.head())

```

# gender distribution

```python

plt.figure(figsize=(5,5))
ax = sns.countplot(data =df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that: the number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb ,annot=True)
plt.title("Relationship between Parents Education and Student's score")
plt.show()
```

```python
#from the above chart we  have concluded that the education of the parents have a good impact on their scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1 ,annot=True)
plt.title("Relationship between Parents Marital Status and Student's Score")
plt.show()
```

```python
#from the above chart we  have concluded that the marital status of the parents have no/negligible impact on their scores
```

```python
gb2 = df.groupby("PracticeSport").agg({"MathScore":'mean', "ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb2 ,annot=True)
plt.title("Relationship between Students practicing sport and Student's Score")
plt.show()
```

```python
#from the above chart we  have concluded that the students practicing sports have  no/negligible impact on their scores
```

# Distribution of Ethnic Group

```python
groupA = df.loc[(df['EthnicGroup']== "group A")].count()
groupB = df.loc[(df['EthnicGroup']== "group B")].count()
groupC = df.loc[(df['EthnicGroup']== "group C")].count()
groupD = df.loc[(df['EthnicGroup']== "group D")].count()
groupE = df.loc[(df['EthnicGroup']== "group E")].count()

l= ["group A", "group B", "group C", "group D" , "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels=l , autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
